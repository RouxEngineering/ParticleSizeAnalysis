### File to Process PyImage CSV files


```python
import pandas as pd 
import os

# create function to take list of csv files and then calculate average for each file and append as new line in dataframe, return concatenated dataframe
def calculate_average(folder_path):
    '''take liste of csv files and calculate averages and append to new dataframe
    
    return: dataframe
    '''
    df_averages = []
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        df = pd.read_csv(file_path)

        filtered_df = df[(df['Feret'] >= 40) & (df['Feret'] <= 110)]
            # Calculate the mean of each column in the filtered DataFrame
            
        column_averages = filtered_df.mean()
        column_averages['File Name'] = file_name
        df_averages.append(column_averages)


    df_averages = pd.concat(df_averages, axis=1).T.reset_index(drop=True)
    columns = ['File Name'] + [col for col in df_averages.columns if col != 'File Name']
    df_averages = df_averages[columns]

    return df_averages
```

```python
folder_path = '/Users/sarah/Documents/areospace research/material recycling project/PyImageJ-Particle-Analysis/Ball Milling Powder Samples/Ball Milling Powder Sample CSV'
df_averages = calculate_average(folder_path)
```
