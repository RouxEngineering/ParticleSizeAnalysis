```python
# import libraries
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.autolimit_mode'] = 'round_numbers'
plt.rcParams['axes.xmargin'] = 0
plt.rcParams['axes.ymargin'] = 0
```

```python
optical_images = pd.read_csv('/Users/sarah/Documents/areospace research/material recycling project/PyImageJ-Particle-Analysis/Optical Images/Optical Images Results/concatenated_output.csv')

optical_images_circularity_counts = optical_images['Circ.'].value_counts().sort_index()
optical_images_circularity_counts = optical_images_circularity_counts.cumsum()
optical_circularity_percentage = (optical_images_circularity_counts / optical_images_circularity_counts.max()) * 100

plt.figure(figsize=(10, 6))
plt.plot(optical_circularity_percentage.index, optical_circularity_percentage.values, linestyle='-', color='black', label='Optical Image Data')

plt.legend()
plt.xlabel('Circularity (unitless)')
plt.ylabel('Cumulative Particle Percentage (%)')
plt.title('Cumulative Particle Percentage (%) vs. Circularity (unitless) For Optical Images')

plt.show()
```

```python
optical_images_feret_counts = optical_images['Feret'].value_counts().sort_index()
optical_images_feret_counts = optical_images_feret_counts.cumsum()
optical_images_percentage = (optical_images_feret_counts / optical_images_feret_counts.max()) * 100

plt.figure(figsize=(10, 6))
plt.plot(optical_images_percentage.index, optical_images_percentage.values, linestyle='-', color='black', label='Optical Image Data')

plt.legend()
plt.xlabel('Feret Diameter (μm)')
plt.ylabel('Cumulative Particle Percentage (%)')
plt.title('Cumulative Particle Percentage (%) vs. Feret Diameter (μm) for Optical Images')

plt.show()
```
