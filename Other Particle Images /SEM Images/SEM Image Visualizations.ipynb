```python
# import libraries
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.autolimit_mode'] = 'round_numbers'
plt.rcParams['axes.xmargin'] = 0
plt.rcParams['axes.ymargin'] = 0
```

```python
# read in different lot data 
lot222r2 = pd.read_csv('/Users/sarah/Documents/areospace research/material recycling project/PyImageJ-Particle-Analysis/SEM Images/Lot 222 R2/Lot222R2 Results/concatenated_output.csv')
lot1440r5 = pd.read_csv('/Users/sarah/Documents/areospace research/material recycling project/PyImageJ-Particle-Analysis/SEM Images/Lot 1440 R5/Lot1440R5 Results/concatenated_output.csv')
lot1440r7 = pd.read_csv('/Users/sarah/Documents/areospace research/material recycling project/PyImageJ-Particle-Analysis/SEM Images/Lot 1440 R7/Lot1440R7 Results/concatenated_output.csv')
lot1440r8 = pd.read_csv('/Users/sarah/Documents/areospace research/material recycling project/PyImageJ-Particle-Analysis/SEM Images/Lot 1440 R8/Lot1440R8 Results/concatenated_output.csv')
```

```python
# calculate percentages and counts for SEM data
lot222r2_circularity_counts = lot222r2['Circ.'].value_counts().sort_index()
lot222r2_circularity_counts = lot222r2_circularity_counts.cumsum()
lot222r2_circularity_percentage = (lot222r2_circularity_counts / lot222r2_circularity_counts.max()) * 100

lot1440r5_circularity_counts = lot1440r5['Circ.'].value_counts().sort_index()
lot1440r5_circularity_counts = lot1440r5_circularity_counts.cumsum()
lot1440r5_circularity_percentage = (lot1440r5_circularity_counts / lot1440r5_circularity_counts.max()) * 100

lot1440r7_circularity_counts = lot1440r7['Circ.'].value_counts().sort_index()
lot1440r7_circularity_counts = lot1440r7_circularity_counts.cumsum()
lot1440r7_circularity_percentage = (lot1440r7_circularity_counts / lot1440r7_circularity_counts.max()) * 100

lot1440r8_circularity_counts = lot1440r8['Circ.'].value_counts().sort_index()
lot1440r8_circularity_counts = lot1440r8_circularity_counts.cumsum()
lot1440r8_circularity_percentage = (lot1440r8_circularity_counts / lot1440r8_circularity_counts.max()) * 100

# plot
plt.figure(figsize=(10, 6))
plt.scatter(lot222r2_circularity_percentage.index, lot222r2_circularity_percentage.values, color='black', label='SEM Lot 222 R2 Data', s=5)
plt.scatter(lot1440r5_circularity_percentage.index, lot1440r5_circularity_percentage.values,  color='blue', label='SEM Lot 1440 R5 Data', s=5)
plt.scatter(lot1440r7_circularity_percentage.index, lot1440r7_circularity_percentage.values, color='red', label='SEM Lot 1440 R7 Data', s=5)
plt.scatter(lot1440r8_circularity_percentage.index, lot1440r8_circularity_percentage.values, color='green', label='SEM Lot 1440 R8 Data', s=5)

plt.legend()
plt.xlabel('Circularity (unitless)')
plt.ylabel('Cumulative Particle Percentage (%)')
plt.title('Cumulative Particle Percentage (%) vs. Circularity (unitless) For SEM Images')

plt.show()
```

```python
# calculate percentages and counts for SEM data
lot222r2_feret_counts = lot222r2['Feret'].value_counts().sort_index()
lot222r2_feret_counts = lot222r2_feret_counts.cumsum()
lot222r2_feret_percentage = (lot222r2_feret_counts / lot222r2_feret_counts.max()) * 100

lot1440r5_feret_counts = lot1440r5['Feret'].value_counts().sort_index()
lot1440r5_feret_counts = lot1440r5_feret_counts.cumsum()
lot1440r5_feret_percentage = (lot1440r5_feret_counts / lot1440r5_feret_counts.max()) * 100

lot1440r7_feret_counts = lot1440r7['Feret'].value_counts().sort_index()
lot1440r7_feret_counts = lot1440r7_feret_counts.cumsum()
lot1440r7_feret_percentage = (lot1440r7_feret_counts / lot1440r7_feret_counts.max()) * 100

lot1440r8_feret_counts = lot1440r8['Feret'].value_counts().sort_index()
lot1440r8_feret_counts = lot1440r8_feret_counts.cumsum()
lot1440r8_feret_percentage = (lot1440r8_feret_counts / lot1440r8_feret_counts.max()) * 100

# plot
plt.figure(figsize=(10, 6))


plt.scatter(lot222r2_feret_percentage.index, lot222r2_feret_percentage.values, color='black', label='SEM Lot 222 R2 Data', s=5)
plt.scatter(lot1440r5_feret_percentage.index, lot1440r5_feret_percentage.values, color='blue', label='SEM Lot 1440 R5 Data', s=5)
plt.scatter(lot1440r7_feret_percentage.index, lot1440r7_feret_percentage.values, color='red', label='SEM Lot 1440 R7 Data', s=5)
plt.scatter(lot1440r8_feret_percentage.index, lot1440r8_feret_percentage.values, color='green', label='SEM Lot 1440 R8 Data', s=5)


plt.legend()
plt.xlabel('Feret Diameter (m)')
plt.ylabel('Cumulative Particle Percentage (%)')
plt.title('Cumulative Particle Percentage (%) vs. Feret Diameter (m) for SEM Images')

plt.show()
```
